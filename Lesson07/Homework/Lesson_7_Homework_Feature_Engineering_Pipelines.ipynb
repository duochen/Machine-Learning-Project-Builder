### Reusable Pipeline Function

This function builds a full preprocessing and modeling pipeline. Run it as-is.

````python
def build_logistic_regression_pipeline(numerical_features, categorical_features):
    # Numerical preprocessing:
    # Fill missing values with the median, then scale the values.
    numeric_transformer = Pipeline(steps=[
        ("imputer", SimpleImputer(strategy="median")),
        ("scaler", StandardScaler())
    ])

    # Categorical preprocessing:
    # Fill missing values with the most common category, then one-hot encode.
    categorical_transformer = Pipeline(steps=[
        ("imputer", SimpleImputer(strategy="most_frequent")),
        ("onehot", OneHotEncoder(handle_unknown="ignore"))
    ])

    # Apply numerical preprocessing to numerical columns
    # and categorical preprocessing to categorical columns.
    preprocessor = ColumnTransformer(
        transformers=[
            ("num", numeric_transformer, numerical_features),
            ("cat", categorical_transformer, categorical_features)
        ]
    )

    # Create the full pipeline:
    # Step 1: preprocess data
    # Step 2: train logistic regression
    model_pipeline = Pipeline(steps=[
        ("preprocessor", preprocessor),
        ("model", LogisticRegression(random_state=42, max_iter=1000))
    ])

    return model_pipeline
````

### Train and Evaluate the Baseline Model

````python
# Split original data into training and test sets
X_train_original, X_test_original, y_train, y_test = train_test_split(
    X_original,
    y,
    test_size=0.25,
    random_state=42,
    stratify=y
)

# Create the baseline pipeline
baseline_pipeline = build_logistic_regression_pipeline(
    numerical_features_original,
    categorical_features
)

# Train the baseline model
baseline_pipeline.fit(X_train_original, y_train)

# Make predictions on the test set
baseline_predictions = baseline_pipeline.predict(X_test_original)

# Calculate accuracy
baseline_accuracy = accuracy_score(y_test, baseline_predictions)

# Display results
print("Baseline Accuracy:", round(baseline_accuracy, 3))
print("\nBaseline Confusion Matrix:")
print(confusion_matrix(y_test, baseline_predictions))
print("\nBaseline Classification Report:")
print(classification_report(y_test, baseline_predictions))
````
