# Part 1: Background

Machine learning data is often messy. Real-world datasets may include:

- Missing values
- Text categories
- Numerical columns with different scales
- Raw features that do not directly show useful patterns

**Feature engineering** means creating better input features from existing data.

Example:

If we have:

- `study_hours`
- `practice_tests`

We can create:

```python
preparation_score = study_hours + practice_tests * 2
```

This new feature may better represent how prepared a student is.

A **pipeline** is a sequence of steps. In this notebook, our pipeline will do:

1. Fill missing values.
2. Scale numerical features.
3. Convert categorical features into numbers.
4. Train a machine learning model.

## Identify Numerical and Categorical Features

A pipeline needs to know which columns are numerical and which are categorical.

````python
# List the numerical features.
numerical_features = [
    "study_hours",
    "sleep_hours",
    "attendance_rate",
    "practice_tests",
    "study_sleep_balance",
    "preparation_score"
]

# List the categorical features.
categorical_features = [
    "internet_access",
    "parent_support",
    "activity_level"
]

print("Numerical features:", numerical_features)
print("Categorical features:", categorical_features)
````

## Teacher Solution: Preprocessing Pipeline

````python
# Numerical preprocessing pipeline:
# 1. Fill missing numerical values with the median.
# 2. Standardize numerical values.

numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])


# Categorical preprocessing pipeline:
# 1. Fill missing categorical values with the most frequent category.
# 2. Convert text categories into numeric one-hot encoded columns.

categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])


# Combine the numerical and categorical preprocessing.
preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numerical_features),
        ("cat", categorical_transformer, categorical_features)
    ]
)
````

# Part 5: Train the Model

Now we will build a full machine learning pipeline.

The full pipeline will include:

1. Preprocessing
2. Logistic regression model

We will also split data into:

- Training set: used to teach the model
- Test set: used to evaluate the model on data it has not seen before

````python
# Create the full pipeline.
# The first step preprocesses the data.
# The second step trains the machine learning model.

model_pipeline = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("model", LogisticRegression(random_state=42))
])
````

````python
# Split the dataset into training and test sets.

X_train, X_test, y_train, y_test = train_test_split(
    X,                   # Input features
    y,                   # Target labels
    test_size=0.25,      # 25% of data is used for testing
    random_state=42,     # Makes the split reproducible
    stratify=y           # Keeps class proportions similar in train and test
)

print("Training rows:", X_train.shape[0])
print("Test rows:", X_test.shape[0])
````

````python
# Train the model pipeline.
# This single line automatically preprocesses the data and trains the model.

model_pipeline.fit(X_train, y_train)
````

# Part 6: Evaluate the Model

After training, we use the test set to evaluate performance.

We will calculate:

1. Accuracy
2. Confusion matrix
3. Classification report

````python
# Use the trained pipeline to make predictions on the test set.
y_pred = model_pipeline.predict(X_test)

# Calculate accuracy.
accuracy = accuracy_score(y_test, y_pred)

print("Model Accuracy:", round(accuracy, 2))
````

````python
# Show the confusion matrix.
# Rows are true labels.
# Columns are predicted labels.

cm = confusion_matrix(y_test, y_pred)

print("Confusion Matrix:")
print(cm)
````

````python
# Show precision, recall, and F1-score.
print("Classification Report:")
print(classification_report(y_test, y_pred))
````

# Part 7: Improve the Model

Now we will compare two versions:

1. **Basic model:** uses only original features.
2. **Improved model:** uses original features plus engineered features.

This helps students understand whether feature engineering can improve a model.

````python
# Recreate the original feature set without engineered features.
X_basic = df.drop("passed_exam", axis=1)

# Define original numerical features.
basic_numerical_features = [
    "study_hours",
    "sleep_hours",
    "attendance_rate",
    "practice_tests"
]

# Define original categorical features.
basic_categorical_features = [
    "internet_access",
    "parent_support",
    "activity_level"
]


# Create preprocessing for the basic model.
basic_preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, basic_numerical_features),
        ("cat", categorical_transformer, basic_categorical_features)
    ]
)


# Create the basic model pipeline.
basic_pipeline = Pipeline(steps=[
    ("preprocessor", basic_preprocessor),
    ("model", LogisticRegression(random_state=42))
])
````

````python
# Split the basic feature data.
X_basic_train, X_basic_test, y_basic_train, y_basic_test = train_test_split(
    X_basic,
    y,
    test_size=0.25,
    random_state=42,
    stratify=y
)

# Train the basic model.
basic_pipeline.fit(X_basic_train, y_basic_train)

# Predict using the basic model.
basic_pred = basic_pipeline.predict(X_basic_test)

# Evaluate the basic model.
basic_accuracy = accuracy_score(y_basic_test, basic_pred)

print("Basic Model Accuracy:", round(basic_accuracy, 2))
print("Improved Model Accuracy:", round(accuracy, 2))
````

````python
# Create a simple bar chart to compare model accuracy.

model_names = ["Basic Features", "Engineered Features"]
model_scores = [basic_accuracy, accuracy]

plt.figure(figsize=(7, 5))
plt.bar(model_names, model_scores)
plt.ylim(0, 1)
plt.ylabel("Accuracy")
plt.title("Model Accuracy Comparison")
plt.show()
````
