# Part 6: Build the Preprocessing Pipeline

We will create two preprocessing paths:

## Numerical data

Missing numerical values will be filled using the **median**.

## Categorical data

Missing categorical values will be filled using the **most frequent value**.

Then text categories will be converted into numbers using **one-hot encoding**.


## Student TODO 9: Create the Preprocessing Steps

````python
# TODO: Create a numerical transformer.
# It should fill missing values using the median.

# numerical_transformer = ...


# TODO: Create a categorical transformer.
# It should fill missing values using the most frequent value.
# It should then apply one-hot encoding.

# categorical_transformer = Pipeline(steps=[
#     ("imputer", ...),
#     ("onehot", ...)
# ])


# TODO: Combine the transformers using ColumnTransformer.

# preprocessor = ColumnTransformer(
#     transformers=[
#         ("num", numerical_transformer, numerical_features),
#         ("cat", categorical_transformer, categorical_features)
#     ]
# )

````

## Teacher Solution 9

````python
# Teacher solution: Create a transformer for numerical columns.
# Missing numerical values will be replaced with the median.
numerical_transformer = SimpleImputer(strategy="median")

# Teacher solution: Create a transformer for categorical columns.
# Step 1: Fill missing categorical values with the most frequent value.
# Step 2: Convert categories into numbers using one-hot encoding.
categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

# Teacher solution: Combine preprocessing for numerical and categorical columns.
preprocessor = ColumnTransformer(
    transformers=[
        ("num", numerical_transformer, numerical_features),
        ("cat", categorical_transformer, categorical_features)
    ]
)

print("Preprocessing pipeline created successfully.")
````

## Teacher Solution 10

````python
# Teacher solution: Split the data into training and test sets.
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42
)

# Print the shapes so we can see how many rows are in each set.
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
````

# Part 8: Train the Logistic Regression Model

Now we will create a full machine learning pipeline.

The pipeline will:

2. Train a Logistic Regression model.

Logistic Regression is a classification model. It is useful when the target is a category such as:

- `1` = passed


## Student TODO 11: Create and Train the Model

````python
# TODO: Create a pipeline with preprocessor and LogisticRegression.

# logistic_model = Pipeline(steps=[
#     ("preprocessor", ...),
#     ("classifier", ...)
# ])


# TODO: Train the model.

# logistic_model.fit(..., ...)

````

## Teacher Solution 11

````python
# Teacher solution: Create the Logistic Regression pipeline.
logistic_model = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("classifier", LogisticRegression())
])

# Teacher solution: Train the model on the training data.
logistic_model.fit(X_train, y_train)

print("Logistic Regression model training complete.")
````

## Student TODO 16: Add `sleep_hours` and Retrain the Model

````python
# TODO: Make a copy of df.

# df_sleep = ...


# TODO: Add a sleep_hours column.

# df_sleep["sleep_hours"] = [5, 6, 7, 5, 8, 6, 8, 7, 8, 6, 7, 9]


# TODO: Create new X and y.

# X_sleep = ...
# y_sleep = ...


# TODO: Update numerical features.

# numerical_features_sleep = ...


# TODO: Build a new preprocessor.

# preprocessor_sleep = ...


# TODO: Split the new data.

# X_train_sleep, X_test_sleep, y_train_sleep, y_test_sleep = ...


# TODO: Create, train, predict, and evaluate a model.

````

## Teacher Solution 16

````python
# Teacher solution: Make a copy of the original dataset.
df_sleep = df.copy()

# Teacher solution: Add a new feature called sleep_hours.
df_sleep["sleep_hours"] = [5, 6, 7, 5, 8, 6, 8, 7, 8, 6, 7, 9]

# Teacher solution: Create new features and target.
X_sleep = df_sleep.drop("passed", axis=1)
y_sleep = df_sleep["passed"]

# Teacher solution: Update the numerical features list.
numerical_features_sleep = ["study_hours", "attendance", "homework_completion", "sleep_hours"]

# The categorical features stay the same.
categorical_features_sleep = ["has_laptop", "study_environment"]

# Teacher solution: Create a new preprocessor for the updated feature list.
preprocessor_sleep = ColumnTransformer(
    transformers=[
        ("num", numerical_transformer, numerical_features_sleep),
        ("cat", categorical_transformer, categorical_features_sleep)
    ]
)

# Teacher solution: Split the updated dataset.
X_train_sleep, X_test_sleep, y_train_sleep, y_test_sleep = train_test_split(
    X_sleep,
    y_sleep,
    test_size=0.3,
    random_state=42
)

# Teacher solution: Create a new Logistic Regression pipeline.
sleep_model = Pipeline(steps=[
    ("preprocessor", preprocessor_sleep),
    ("classifier", LogisticRegression())
])

# Teacher solution: Train the model.
sleep_model.fit(X_train_sleep, y_train_sleep)

# Teacher solution: Make predictions.
sleep_pred = sleep_model.predict(X_test_sleep)

# Teacher solution: Calculate accuracy.
sleep_accuracy = accuracy_score(y_test_sleep, sleep_pred)

print("Original Logistic Regression Accuracy:", logistic_accuracy)
print("Accuracy After Adding Sleep Hours:", sleep_accuracy)
````
