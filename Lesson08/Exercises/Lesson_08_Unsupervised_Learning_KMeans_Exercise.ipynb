# Part 1: Background

## What is unsupervised learning?

In supervised learning, we train a model using input features **X** and known answers **y**.

Examples:

- Predict house price: `y = price`
- Classify an email: `y = spam or not spam`
- Predict disease risk: `y = high risk or low risk`

In **unsupervised learning**, we do not have known answers. We only have input features.

The goal is to discover hidden patterns.

## What is clustering?

Clustering means grouping similar data points together.

Example:

A mall may not know customer types in advance, but it may have data about:

- Annual income
- Spending score

A clustering algorithm may discover groups such as:

- Low income, high spending
- Medium income, medium spending
- High income, low spending
- High income, high spending

## What is K-Means?

K-Means is a popular clustering algorithm.

The letter **K** means the number of clusters you want the model to find.

For example:

```python
KMeans(n_clusters=3)
```

means:

> Please divide the data into 3 groups.

Important: K-Means does not know the “true” names of the groups. It only assigns cluster numbers such as 0, 1, and 2.

# Part 6: Evaluate the Model

Evaluating clustering is different from evaluating classification.

In classification, we can compare predictions with known labels.

In clustering, we usually do not have known labels.

So we use other methods, such as:

1. Visual inspection
2. Cluster summaries
3. Inertia
4. Silhouette score

## What is inertia?

Inertia measures how close points are to their cluster centers.

Lower inertia usually means points are closer to their centers.

However, lower inertia does not always mean better clustering, because adding more clusters usually lowers inertia.

## What is silhouette score?

Silhouette score measures how well-separated the clusters are.

It usually ranges from -1 to 1.

A higher score is usually better.

````python
# ============================================================
# Part 6 Code: Summarize clusters
# ============================================================

# Group the dataframe by cluster and calculate the average values.
cluster_summary = df.groupby("Cluster").mean()

# Show the average income and spending score for each cluster.
cluster_summary
````

````python
# ============================================================
# Part 6 Code: Print model evaluation values
# ============================================================

# Inertia is stored inside the trained KMeans model.
print("Inertia:")
print(kmeans.inertia_)

# Silhouette score compares how similar a point is to its own cluster
# compared with other clusters.
score = silhouette_score(X_scaled, labels)

print("\nSilhouette Score:")
print(score)
````
