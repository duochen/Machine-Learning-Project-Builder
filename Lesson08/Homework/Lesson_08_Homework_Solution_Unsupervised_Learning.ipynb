# Part 1: Background

## What is unsupervised learning?

In **supervised learning**, the dataset includes known answers, also called labels.

Examples:

- Predicting a house price
- Classifying whether an email is spam
- Predicting whether a student passes or fails

In **unsupervised learning**, the dataset does **not** include known answers.

The model tries to discover hidden patterns by itself.

## What is clustering?

**Clustering** means grouping similar data points together.

In this homework, we will group students based on:

1. Weekly study hours
2. Number of practice problems completed per week

The dataset does not tell us student types in advance.  
K-Means will discover possible groups.

# Part 10: Use the Elbow Method

We started with `K = 3`, but was that a good choice?

The **Elbow Method** helps us compare different K values.

Steps:

1. Train K-Means for several K values.
2. Save the inertia for each K.
3. Plot K versus inertia.
4. Look for the bend or “elbow” in the curve.

````python
# ============================================================
# Part 10 Student TODO: Elbow Method
# ============================================================

# TODO: Create an empty list to store inertia values.

inertias = None

# TODO: Create K values from 1 to 8.
# Hint: use range(1, 9)

K_values = None

# TODO: Write a loop to train KMeans for each K.
# For each K:
# 1. Create a KMeans model
# 2. Fit it on X_scaled
# 3. Append model.inertia_ to inertias

# for k in K_values:
#     model = ...
#     model.fit(...)
#     inertias.append(...)

# TODO: Create the elbow plot.
plt.figure(figsize=(7, 5))
plt.plot(None, None, marker="o")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.title("Elbow Method for Choosing K")
plt.show()
````

## Teacher Solution: Elbow Method

````python
# ============================================================
# Part 10 Teacher Solution: Elbow Method
# ============================================================

# Create an empty list to store inertia values.
inertias = []

# Try K values from 1 through 8.
K_values = range(1, 9)

# Train one K-Means model for each K value.
for k in K_values:
    # Create the K-Means model for the current K value.
    model = KMeans(n_clusters=k, random_state=42, n_init=10)

    # Train the model using scaled data.
    model.fit(X_scaled)

    # Save the inertia value.
    inertias.append(model.inertia_)

# Create a figure.
plt.figure(figsize=(7, 5))

# Plot K values against inertia values.
plt.plot(K_values, inertias, marker="o")

# Add axis labels and title.
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.title("Elbow Method for Choosing K")

# Show the plot.
plt.show()
````
