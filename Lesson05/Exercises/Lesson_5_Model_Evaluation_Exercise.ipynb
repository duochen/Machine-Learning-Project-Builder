# Part 1: Background

Machine learning models are not useful just because they can make predictions.

A strong machine learning project must also prove:

- How often the model is correct
- What kinds of mistakes the model makes
- Whether the model performs well on new data
- Whether one model is better than another model

In this lesson, we will use a **classification problem**.

A classification model predicts a category, such as:

- spam or not spam
- pass or fail
- disease or no disease
- malignant or benign

For this notebook, we will use a small medical-related dataset built into scikit-learn.

Important note:

> This notebook is for education only. It should not be used for real medical diagnosis.

## Key Terms

| Term | Simple Meaning |
|---|---|
| Training data | Data the model learns from |
| Test data | New data used to check the model |
| Prediction | The model's answer |
| Accuracy | Percent of predictions that are correct |
| Confusion matrix | A table that shows correct and incorrect predictions |
| Precision | When the model predicts a class, how often is it correct? |
| Recall | Out of all real examples of a class, how many did the model find? |
| F1-score | A balance between precision and recall |

## How to Read the Confusion Matrix

For this dataset:

- `0 = malignant`
- `1 = benign`

The confusion matrix has this general format:

| | Predicted Malignant | Predicted Benign |
|---|---:|---:|
| Actually Malignant | Correct malignant predictions | Malignant cases missed |
| Actually Benign | Benign cases wrongly flagged | Correct benign predictions |

A medical-related model should be evaluated carefully because different mistakes may have different consequences.
