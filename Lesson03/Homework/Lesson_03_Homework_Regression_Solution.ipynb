````python
# ============================================================
# Part 2: Create the Dataset
# Teacher Solution
# ============================================================

# Create a small plant growth dataset.
# Each list represents one column in the dataset.
data = {
    "sunlight_hours": [4, 5, 6, 7, 8, 5, 6, 7, 8, 9,
                       4, 5, 6, 7, 8, 9, 6, 7, 8, 9],

    "water_ml": [100, 120, 150, 180, 200, 110, 140, 170, 210, 230,
                 90, 130, 160, 190, 220, 240, 150, 175, 205, 235],

    "fertilizer_grams": [1, 1, 2, 2, 3, 1, 2, 2, 3, 3,
                         0, 1, 2, 2, 3, 4, 2, 2, 3, 4],

    "plant_height_cm": [8, 10, 13, 16, 19, 9, 12, 15, 20, 23,
                        7, 11, 14, 17, 21, 25, 13, 16, 20, 24]
}

# Convert the dictionary into a pandas DataFrame.
df = pd.DataFrame(data)

# Display the first five rows of the dataset.
df.head()
````

## Teacher Solution — Part 3

# Part 4: Prepare Features and Target

Machine learning models need:

- `X`: the input features
- `y`: the target we want to predict

For this project:

```text
X = sunlight_hours, water_ml, fertilizer_grams
y = plant_height_cm
```

Then we split the data into:

- training data
- test data

The model learns from the training data and is evaluated on the test data.


````python
# ============================================================
# Part 4: Prepare Features and Target
# Student TODO Section
# ============================================================

# TODO 1: Create X using the three feature columns.
# X = df[[...]]


# TODO 2: Create y using the target column.
# y = df[...]


# TODO 3: Split the data using train_test_split.
# Use test_size=0.25 and random_state=42.


# TODO 4: Print the shapes of X_train, X_test, y_train, and y_test.
````

## Teacher Solution — Part 4

````python
# ============================================================
# Part 4: Prepare Features and Target
# Teacher Solution
# ============================================================

# Select the input features.
# These are the columns the model will use to make predictions.
X = df[["sunlight_hours", "water_ml", "fertilizer_grams"]]

# Select the target.
# This is the number the model will try to predict.
y = df["plant_height_cm"]

# Split the dataset into training data and test data.
# 25% of the data will be used for testing.
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=42
)

# Print the shapes to confirm the split worked correctly.
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
````

# Part 5: Train the Linear Regression Model

Now we will train a linear regression model.

Linear regression tries to learn a formula like:

```text
plant_height ≈ a × sunlight_hours + b × water_ml + c × fertilizer_grams + intercept
```

The model learns the values of `a`, `b`, `c`, and the intercept from the training data.


````python
# ============================================================
# Part 5: Train the Model
# Student TODO Section
# ============================================================

# TODO 1: Create a LinearRegression model.
# model = ...


# TODO 2: Train the model using X_train and y_train.
# model.fit(...)


# TODO 3: Print a success message.
````

## Teacher Solution — Part 5

````python
# ============================================================
# Part 5: Train the Model
# Teacher Solution
# ============================================================

# Create a linear regression model.
model = LinearRegression()

# Train the model using the training data.
model.fit(X_train, y_train)

# Print a success message.
print("Linear regression model trained successfully.")
````

# Part 6: Make Predictions

After the model is trained, we can use it to predict plant height for the test data.

The test data contains examples the model did not use during training.


````python
# ============================================================
# Part 6: Make Predictions
# Student TODO Section
# ============================================================

# TODO 1: Use the model to predict plant height for X_test.
# y_pred = ...


# TODO 2: Create a table comparing actual and predicted values.
# comparison = pd.DataFrame({...})


# TODO 3: Display the comparison table.
````

## Teacher Solution — Part 6

````python
# ============================================================
# Part 6: Make Predictions
# Teacher Solution
# ============================================================

# Use the trained model to predict plant height for the test data.
y_pred = model.predict(X_test)

# Create a comparison table.
comparison = pd.DataFrame({
    "Actual Plant Height": y_test.values,
    "Predicted Plant Height": np.round(y_pred, 2),
    "Prediction Error": np.round(y_test.values - y_pred, 2)
})

# Display the comparison table.
display(comparison)
````

# Part 7: Evaluate the Model

We will use four common regression metrics.

| Metric | Meaning |
|---|---|
| MAE | Average absolute error |
| MSE | Average squared error |
| RMSE | Typical prediction error in the same unit as the target |
| R² | How well the model explains the target |

For this project, RMSE is measured in **centimeters**.


````python
# ============================================================
# Part 7: Evaluate the Model
# Student TODO Section
# ============================================================

# TODO 1: Calculate MAE.
# mae = ...


# TODO 2: Calculate MSE.
# mse = ...


# TODO 3: Calculate RMSE.
# rmse = ...


# TODO 4: Calculate R² score.
# r2 = ...


# TODO 5: Print all four metrics.
````

## Teacher Solution — Part 7

````python
# ============================================================
# Part 7: Evaluate the Model
# Teacher Solution
# ============================================================

# Calculate Mean Absolute Error.
# This tells us the average size of the prediction mistake.
mae = mean_absolute_error(y_test, y_pred)

# Calculate Mean Squared Error.
# This squares each error before averaging.
mse = mean_squared_error(y_test, y_pred)

# Calculate Root Mean Squared Error.
# This puts the error back into the original unit: centimeters.
rmse = np.sqrt(mse)

# Calculate R² score.
# This tells us how well the model explains the target.
r2 = r2_score(y_test, y_pred)

# Print the evaluation results.
print("Model Evaluation Results")
print("------------------------")
print("MAE:", round(mae, 2))
print("MSE:", round(mse, 2))
print("RMSE:", round(rmse, 2))
print("R²:", round(r2, 3))
````

# Part 8: Interpret the Model

Linear regression gives each feature a coefficient.

A coefficient tells us how much the prediction changes when that feature increases by 1 unit, assuming the other features stay the same.

Important note:

> Coefficients show patterns in the data. They do not automatically prove cause and effect.

## Teacher Solution — Part 8

````python
# ============================================================
# Part 8: Interpret the Model
# Teacher Solution
# ============================================================

# Print the model intercept.
print("Intercept:", round(model.intercept_, 2))

# Create a table of coefficients.
coefficients = pd.DataFrame({
    "Feature": X.columns,
    "Coefficient": np.round(model.coef_, 4)
})

# Display the coefficients.
display(coefficients)
````

## Teacher Solution — Optional Challenge

````python
# ============================================================
# Part 11: Optional Challenge
# Teacher Solution
# ============================================================

# Import two additional regression models.
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor

# Create a dictionary of models.
models = {
    "Linear Regression": LinearRegression(),
    "KNN Regression": KNeighborsRegressor(n_neighbors=3),
    "Random Forest Regression": RandomForestRegressor(random_state=42)
}

# Create an empty list to store the results.
model_results = []

# Train and evaluate each model.
for model_name, reg_model in models.items():

    # Train the model on the training data.
    reg_model.fit(X_train, y_train)

    # Predict plant height on the test data.
    predictions = reg_model.predict(X_test)

    # Calculate MAE.
    model_mae = mean_absolute_error(y_test, predictions)

    # Calculate RMSE.
    model_rmse = np.sqrt(mean_squared_error(y_test, predictions))

    # Calculate R² score.
    model_r2 = r2_score(y_test, predictions)

    # Store the results in a dictionary.
    model_results.append({
        "Model": model_name,
        "MAE": round(model_mae, 2),
        "RMSE": round(model_rmse, 2),
        "R²": round(model_r2, 3)
    })

# Convert results into a DataFrame.
model_results_df = pd.DataFrame(model_results)

# Display the model comparison table.
display(model_results_df)
````

# Common Errors and Fixes

| Error | Why It Happens | How to Fix It |
|---|---|---|
| `NameError: name 'pd' is not defined` | pandas was not imported | Run `import pandas as pd` |
| `NameError: train_test_split is not defined` | scikit-learn function was not imported | Run the import cell |
| `ValueError: Expected 2D array` | Feature data was passed as a Series | Use double brackets: `df[["column_name"]]` |
| `ValueError: inconsistent numbers of samples` | `X` and `y` have different lengths | Check that both come from the same DataFrame |
| Negative R² score | Model is performing poorly | Check data quality, features, and train/test split |
| Very high accuracy on tiny data | Dataset may be too simple | Use more realistic data |
