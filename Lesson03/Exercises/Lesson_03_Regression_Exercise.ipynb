# Part 1: Background

## What is regression?

Regression is a type of supervised machine learning. It is used when the target we want to predict is a **number**.

Examples:

| Question | Target |
|---|---|
| What will tomorrow's temperature be? | Temperature |
| What will a house sell for? | Price |
| What exam score might a student get? | Score |
| How much pollution will be in the air tomorrow? | PM2.5 level |
| How tall will a plant grow after 30 days? | Height |

In this notebook, our question is:

> Can we predict a student's exam score using study hours, sleep hours, and practice tests?


## Regression Workflow

Most regression projects follow this workflow:

1. Load the dataset.
2. Explore the data.
3. Separate features `X` and target `y`.
4. Split data into training and test sets.
5. Train a regression model.
6. Make predictions.
7. Evaluate the model.
8. Improve the model.

## Interpret the Model

Linear regression gives each feature a **coefficient**.

A positive coefficient means that when the feature increases, the prediction usually increases, assuming the other features stay the same.

A negative coefficient means that when the feature increases, the prediction usually decreases, assuming the other features stay the same.

# Common Errors and Fixes

| Error | Why it happens | How to fix it |
|---|---|---|
| `NameError: name 'pd' is not defined` | pandas was not imported | Run `import pandas as pd` |
| `NameError: train_test_split is not defined` | scikit-learn function was not imported | Run the import cell again |
| `ValueError: Expected 2D array` | Feature data was passed as a Series | Use double brackets: `df[["hours_studied"]]` |
| Poor model score | Dataset is very small | Add more realistic data |
| Negative R² score | Model performs worse than predicting the average | Check features and train-test split |
| Students confuse regression and classification | They focus on the algorithm name | Regression predicts numbers; classification predicts categories |
