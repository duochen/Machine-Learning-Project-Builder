A **Random Forest** combines many decision trees.

For classification:

> Tree 1 predicts class 0  
> Tree 2 predicts class 1  
> Tree 3 predicts class 1  
> Final prediction: class 1

Random forests often perform better than one decision tree because they reduce the risk of relying on one weak or overfitted tree.
