# Part 1: Background

## What is the Digits dataset?

The scikit-learn **Digits dataset** contains small grayscale images of handwritten digits from 0 to 9.

Each image is:

- 8 pixels wide
- 8 pixels tall
- represented as 64 numeric features
- labeled with the correct digit

## What is image classification?

**Image classification** means giving an image a label.

Examples:

- A picture of a handwritten `7` should be classified as digit `7`.
- A leaf image might be classified as healthy or diseased.
- A camera-trap image might be classified as deer, fox, bird, or empty.

## What is a neural network?

A **neural network** is a machine learning model that learns patterns from data.

In this homework, you will use `MLPClassifier`, which is a simple feedforward neural network from scikit-learn.
